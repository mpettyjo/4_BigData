```python
'''
This program is used to make the final datafile for the 
BigData1 paper.
'''

# import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import corner
from matplotlib.patches import Patch

from tabulate import tabulate
```

```python
df_all = pd.read_csv('/Users/z5380625/Documents/Research/4_BigData/B971_def2TZVPD_DataFile1.csv') 
df_bio = pd.read_csv('/Users/z5380625/Documents/Research/4_BigData/6_Atoms/biosig_CHNOPS.csv') 

# df_bio[df_bio['Formula_ID']=='C6H8_50']
```

```python
df_ground=df_all[df_all['Relative Energy [kJ/mol]']==0]
formula_ID_list = df_ground['Formula'].values
IUPAC_chemname_list = []
formula_list = []
for formula_ID in formula_ID_list:
    IUPAC_chemname = df_bio[df_bio['Formula_ID']==formula_ID]['IUPAC_chemname'].values[0]
    formula = df_bio[df_bio['Formula_ID']==formula_ID]['Formula'].values[0]
    IUPAC_chemname_list.append(IUPAC_chemname)
    formula_list.append(formula)
```

```python
df_ground = df_ground.rename({'Formula':'Formula ID','mu_a [D]':'mu_A [D]','mu_b [D]':'mu_B [D]','mu_c [D]':'mu_C [D]'},axis=1)
df_ground.insert(loc=0,column='Formula',value=formula_list)
df_ground.insert(loc=2,column='IUPAC_chemname',value=IUPAC_chemname_list)
df_ground = df_ground.drop(labels=['Formula ID','Conf#','Rotamer Degeneracy','Relative Energy [kJ/mol]','BW Percent at 10 K','BW Percent at 50 K','BW Percent at 100 K','BW Percent at 298.15 K','mu_amu_bmu_c'],axis=1)
df_ground = df_ground.sort_values(by=['# Atoms'])
df_ground = df_ground.reset_index(drop=True)
```

```python
df_ground.to_csv('/Users/z5380625/Documents/Research/4_BigData/B971_def2TZVPD_GroundState.csv')
```
